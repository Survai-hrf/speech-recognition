```python
import whisper
import moviepy.editor as mp
import json
import datetime
#from stable_whisper import modify_model
```

```python
video_id = 'user_interview_04.mp4'

# create file name for audio
audio_file = video_id.split('.')[0]
audio_file = f'{audio_file}.wav'

# split audio from video file
clip = mp.VideoFileClip(video_id)
clip.audio.write_audiofile(audio_file)

# load speech to text model
model = whisper.load_model('medium')
result = model.transcribe(audio_file)
```

```python
with open('user_interview_04.json', 'w+') as file:
    json.dump(result['text'], file)
```

```python
output = {}
segments = []

for i in result['segments']:
    segments.append({
        'start': int(i['start']),
        'text': i['text']
    })

output['uniqueId'] = video_id
output['originalLanguage'] = result['language']
output['transcription'] = result['text']
output['segments'] = segments

output
```
